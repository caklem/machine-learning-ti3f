```python
import pandas as pd # Ini adalah pernyataan impor yang mengimpor pustaka pandas dan memberi alias pd ke pustaka tersebut
import numpy as np # Ini adalah pernyataan impor yang mengimpor pustaka numpy dan memberi alias np ke pustaka tersebut
import matplotlib.pyplot as plt #  Ini adalah pernyataan impor yang mengimpor modul pyplot dari pustaka matplotlib dan memberi alias plt
from sklearn.cluster import KMeans # adalah pernyataan impor yang mengimpor algoritma klastering K-Means dari pustaka Scikit-Learn (sklearn). 

df = pd.read_csv('Iris.csv') #  adalah pernyataan yang membaca data dari file CSV dengan nama 'Iris.csv' dan menyimpannya dalam variabel df

df.head() # digunakan untuk melihat sebagian awal dari data dan memeriksa struktur data tersebut
```

```python
# Seleksi Fitur
# pada X = df.iloc[:, 1:-1]:
# df adalah data frame yang sebelumnya dibaca dari file CSV, dan itu adalah sumber data.
# iloc adalah metode pandas yang digunakan untuk indeks berdasarkan posisi dalam dataframe.
# [:, 1:-1] adalah pemotongan (slicing) data yang dilakukan.
# 1:-1 mengacu pada kolom-kolom dari kolom kedua hingga kolom sebelum kolom terakhir.

# sedangkan pada y = df.iloc[:, -1]:
# [:, -1] mengacu pada kolom terakhir dalam data frame.

X = df.iloc[:, 1:-1] #
y = df.iloc[:, -1]
```

```python
# Plot Data
# Karena data 4 dimensi, maka akan kita coba
# plot cluster berdasarkan Sepal Length dan Sepal Width  saja

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100)
```

```python
# Buat Model KMeans
# Kali ini kita coba menggunakan k=2 - anggap saja kita tidak tahu jumlah label ada 3 :)

from sklearn.cluster import KMeans # adalah pernyataan impor yang mengimpor kelas KMeans 
                                   # dari modul cluster dalam pustaka Scikit-Learn. 

# Inisiasi obyek KMeans
cl_kmeans = KMeans(n_clusters=2) # adalah pernyataan yang menginisialisasi objek cl_kmeans sebagai model K-Means.
                                 # n_clusters=2 mengatur jumlah klaster yang diinginkan. Dalam hal ini, kita ingin mengelompokkan data menjadi 2 klaster.
# Fit dan predict model
y_kmeans = cl_kmeans.fit_predict(X) # adalah pernyataan yang menghitung klastering pada data X menggunakan model K-Means yang telah diinisialisasi sebelumnya.
                                    # fit_predict adalah metode yang digunakan untuk melakukan dua hal sekaligus, yaitu
                                    # 1. Melatih (fit) model K-Means menggunakan data X
                                    # 2. Memprediksi (predict) klaster dari setiap titik data dalam X
```

```python
# Plot hasi cluster berdasarkan Sepal Length dan Sepal Width
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100, c=y_kmeans) 
# adalah pernyataan yang digunakan untuk membuat scatter plot (grafik titik-titik) berdasarkan dua fitur,
# s=100 mengatur ukuran titik pada plot menjadi 100.
# c=y_kmeans mengatur warna titik-titik berdasarkan label klaster yang telah dihasilkan sebelumnya dan disimpan dalam variabel y_kmeans


# Plot centroid
centers = cl_kmeans.cluster_centers_
# adalah pernyataan yang digunakan untuk mengambil koordinat pusat klaster (centroid) yang dihasilkan oleh model K-Means yang telah dibuat sebelumnya.
# cl_kmeans.cluster_centers_ adalah atribut dari objek model 
# cl_kmeans yang berisi koordinat pusat klaster.

plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
# adalah pernyataan yang digunakan untuk membuat scatter plot yang menunjukkan pusat-pusat klaster (centroid).
# centers[:, 0] dan centers[:, 1] mengacu pada koordinat pusat klaster dalam kolom pertama dan kedua.
# c='red' mengatur warna titik-titik pusat klaster menjadi merah.
# s=200 mengatur ukuran titik pusat klaster menjadi 200.
# alpha=0.5 mengatur tingkat transparansi pusat klaster sehingga titik pusat klaster tidak sepenuhnya menutupi titik data.
```

```python
# Cek Nilai SSE
print(f'Nilai SSE: {cl_kmeans.inertia_}')
# f'Nilai SSE: {cl_kmeans.inertia_} adalah sebuah f-string yang menggabungkan teks statis "Nilai SSE: "
# cl_kmeans.inertia_ adalah atribut model K-Means yang berisi nilai SSE.
```

```python
# Implementasi Metode Elbow

# List nilai SSE
sse = [] # adalah pembuatan sebuah daftar kosong yang akan digunakan untuk menyimpan nilai SSE dari berbagai iterasi dengan jumlah klaster yang berbeda.

# Cari k terbaik dari 1-10
K = range(1,10) # adalah pembuatan range (rentang) dari 1 hingga 9 (10 tidak termasuk) yang akan digunakan sebagai kandidat nilai k

# Cek nilai SSE setiap k
for k in K: # perulangan yang akan mengulang dari 1 hingga 9
 kmeanModel = KMeans(n_clusters=k) # perulangan yang akan mengulang dari 1 hingga 9
 kmeanModel.fit(X) # proses pelatihan model K-Means menggunakan data X.
 sse.append(kmeanModel.inertia_) # penambahan nilai SSE dari model K-Means yang telah dilatih ke dalam daftar sse


# Plotting the distortions
plt.figure(figsize=(8,4)) #  pembuatan objek gambar untuk plot dengan ukuran 8x4 inci.
plt.plot(K, sse, "bx-") # pernyataan yang membuat plot garis berdasarkan nilai K
plt.xlabel("k") # penambahan label sumbu X yang menunjukkan nilai k (jumlah klaster).
plt.ylabel("SSE") # penambahan label sumbu Y yang menunjukkan nilai SSE.
plt.title("Metode Elbow untuk Mengetahui Jumlah k Terbaik") # penambahan judul plot.
plt.show() # menampilkan plot ke layar.
```

```python
# Cek Nilai SSE setiap k
for idx, sse_val in enumerate(sse, start=1): 
# perulangan for untuk mengiterasi melalui elemen-elemen dalam daftar sse
# enumerate(sse, start=1) digunakan untuk mengindeks elemen-elemen dalam sse dimulai dari 1. 

    print(f'k={idx}; SSE={sse_val}')
    # f-string yang menggabungkan teks statis dengan nilai k dan nilai SSE.
```
