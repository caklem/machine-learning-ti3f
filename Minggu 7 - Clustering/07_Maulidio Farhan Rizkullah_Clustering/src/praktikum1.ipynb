## KMeans

```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster  import KMeans

df = pd.read_csv('../docs/Iris.csv') # membaca data pada file directory

df.head() # menampilkan data
```

```python
# Menyeleksi Fitur
X = df.iloc[:, 1:-1]
y = df.iloc[:, -1]

X.head() # menampilkan data
```

```python
# plot data
# plot cluster berdasarkan SepalLengthCm dan SepalWidthCm

plt.scatter(X.iloc[:,0], X.iloc[:,1], s=100)
```

```python
# Membuat model KMeans

# inisiasi objek
cl_kmeans = KMeans(n_clusters=2)

# fit dan predict model
y_means = cl_kmeans.fit_predict(X)
```

```python
#plot hasil cluster berdasarkan SepalLengthCm dan SepalWidthCm
plt.scatter(X.iloc[:,0], X.iloc[:,1], s=100, c=y_means)

#plot centroid untuk memberikan titik tengah berwarna merah
centers = cl_kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, alpha=0.5)
```

```python
#mengecek nilai SSE
print(f'Nilai SSE: {cl_kmeans.inertia_}')
```

```python
# implementasi method Elbow

# mengelist nilai SSE
sse = []

# mencari k terbaik dari 1-10
K = range(1,10)

# mengecek nilai SSE dari setiap k
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    sse.append(kmeanModel.inertia_)

# Plotting distortion
plt.figure(figsize=(8,4))
plt.plot(K, sse, "bx-")
plt.xlabel("k")
plt.ylabel("SSE")
plt.title("Metode Elbow Untuk Mengetahui Jumlah k Terbaik")
plt.show()
```

```python
# Mengecek nilai SSE setiap k

for idx, sse_val in enumerate(sse, start=1):
    print(f'k={idx}; SSE={sse_val}')
```
