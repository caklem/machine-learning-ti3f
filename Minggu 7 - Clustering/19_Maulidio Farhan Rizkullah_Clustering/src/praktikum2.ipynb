### Konsep K-Means untuk klasterisasi data

```python
import matplotlib.pyplot as plt
import seaborn as sns;  sns.set()
import numpy as np
```

```python
from sklearn.datasets import make_blobs # import library make_blobs

# membuat datasets buatan dengan 300 sample dan 4 pusat cluster
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# menampilkan plot
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans # import library KMeans

# inisiasi model
kmeans = KMeans(n_clusters=4)

# fit datasets
kmeans.fit(X)

# predict datasets
y_kmeans = kmeans.predict(X)

```

```python
# menampilkan plot
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# plot centroid
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

## Algoritma Expectation-Maximization

```python
from sklearn.metrics import pairwise_distances_argmin # mengimport pairwise_distances_argmin

# membuat function find_clusters
def find_clusters(X, n_clusters, rseed=2):
    # memilih cluster secara acak 
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # input label center yang baru
        labels = pairwise_distances_argmin(X, centers)
        
        # mengupdate center dari titik baru
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # mengecek konvergensi
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

# menjalankan function find_clusters
centers, labels = find_clusters(X, 4)

# menampilkan plot
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
```

```python
# perubahan random
centers, labels = find_clusters(X, 4, rseed=0)

# menampilkan plot
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
```

```python
#mengoptimalisasi jumlah cluster

# membuat variabel labels yang berisi model KMeans beserta fit and predict X
labels = KMeans(6, random_state=0).fit_predict(X)

# menampilkan plot
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

#### Contoh Kasus 1: Karakter Angka

```python
from sklearn.datasets import load_digits # import load_digits

# membuat datasets yang berisi gambar digit tulisan tangan
digits = load_digits()

digits.data.shape
```

```python
# menerapkan kmeans

# inisiasi model KMeans
kmeans = KMeans(n_clusters=10, random_state=0)

# fit dan predict cluster
clusters = kmeans.fit_predict(digits.data)

kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))

centers = kmeans.cluster_centers_.reshape(10, 8, 8)

#melakukan perulangan
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
from scipy.stats import mode #import mode

# inisiasi labels dengan ukuran yang sama seperti clusters
labels = np.zeros_like(clusters)

# melakukan perulangan
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

```python
from sklearn.metrics import accuracy_score # import accuracy_score

accuracy_score(digits.target, labels)
```

```python
from sklearn.metrics import confusion_matrix # import confusion_matrix

# inisiasisi confusion_matrix
mat = confusion_matrix(digits.target, labels)

# menampilkan heatmap
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
from sklearn.manifold import TSNE #import library TSNE dari sklearn

# inisiasi model tsne
tsne = TSNE(n_components=2, init='random', random_state=0)

# fit model
digits_proj = tsne.fit_transform(digits.data)

# menghitung cluster
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits_proj)

# permutasi label
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]

# menghitung akurasi
accuracy_score(digits.target, labels)
```

#### Studi kasus 2: Kompres Citra

```python
from sklearn.datasets import load_sample_image # mengimport load_sample_image

# mengambil sample image flower.jpg dari dataset
flower = load_sample_image("flower.jpg")
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(flower)
```

```python
flower.shape
```

```python
# melakukan normalisasi
data = flower / 255.0

# melakukan reshape
data = data.reshape(427 * 640, 3)
data.shape
```

```python
# membuat function plot_pixels yang digunakan untuk memvisualiasikan distribusi warna dalam citra berdasarkan nilai pixel dalam saluran warna RGB
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    # menampilkan plt
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20)
```

```python
# menggunakan function yang telah dibuat 
plot_pixels(data, title='Input color space: 16 million possible colors')
```

```python
import warnings; warnings.simplefilter('ignore')  # Fix NumPy issues.

from sklearn.cluster import MiniBatchKMeans # import library MiniBatchKMeans

# inisiasi model kmeans
kmeans = MiniBatchKMeans(16)

# fit model kmeans
kmeans.fit(data)

new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

# menampilkan plot dengan function yang sebelumnya telah dibuat
plot_pixels(data, colors=new_colors,title="Reduced color space: 16 colors")
```

```python
# melakukan reshape
flower_recolored = new_colors.reshape(flower.shape)

# menampilkan citra
fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(flower)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(flower_recolored)
ax[1].set_title('16-color Image', size=16)
```
