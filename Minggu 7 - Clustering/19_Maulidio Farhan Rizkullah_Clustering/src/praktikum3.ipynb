#### Self-optimizing Map (SOM)

```python
# Import library
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Load dataset
iris = datasets.load_iris()
data = iris.data
```

```python
# melakukan normalisasi data
data = data/data.max(axis=0)
```

```python
map_size = (10,10)

# inisiasi SOM
som = MiniSom(map_size[0], map_size[1], data.shape[1], sigma=0.5, learning_rate=0.5)
```

```python
# melakukan inisialisasi bobot SOM secara acak
som.random_weights_init(data)
```

```python
# melakukan pelatihan SOM menggunakan train_random
num_epochs = 100
som.train_random(data, num_epochs)
```

```python
# menampilkan visual hasil SOM
plt.figure(figsize=(8, 8))
for i, x in enumerate(data):
    w = som.winner(x)  # Pemenang untuk sampel x
    plt.text(w[0]+.5, w[1]+.5, str(i+1), color='k', fontdict={'weight': 'bold', 'size': 11})
plt.pcolor(som.distance_map().T, cmap='bone_r', alpha=.2)
plt.colorbar()

plt.show()
```
